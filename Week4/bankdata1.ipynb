```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  
import plotly.express as px


# load data
population = pd.read_csv("B:/prepinsta/Week4/country_population.csv")
fertility_rate = pd.read_csv("B:/prepinsta/Week4/fertility_rate.csv")
metadata=pd.read_csv("B:/prepinsta/Week4/Metadata_Country.csv")
life_expectancy = pd.read_csv("B:/prepinsta/Week4/life_expectancy.csv")



```

```python

# List of columns to be removed
columns_to_remove = ["Indicator Name", "Indicator Code"]

# Dropping the specified columns
population = population.drop(columns=columns_to_remove)

# Renaming the 'country Name' column
population.rename(columns={"country Name": "Country Name"}, inplace=True)

# Filtering rows where 'Country Name' is not 'not classified'
population = population[population['Country Name'] != 'not classified']

# Converting column names to lowercase and replacing spaces with underscores
population.columns = population.columns.str.lower().str.replace(' ', '_')

# Filling NaN values with the median of each row for columns starting from the third column
population.iloc[:, 2:] = population.iloc[:, 2:].apply(lambda row: row.fillna(row.median()), axis=1)

population
```
